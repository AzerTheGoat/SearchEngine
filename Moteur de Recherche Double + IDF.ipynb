```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import string
import math
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
with open("DATA/CISI.ALLnettoye", "r") as f:
    contenu = f.read()
    print(contenu)
```

```python
# Rechercher et diviser la chane de donnes par rapport  "I {chiffre}"
data_list = re.split(r'\n(?=.I \d+)', contenu)
data_list = [re.sub(r'^\.I \d+\n', '', elem) for elem in data_list]
print(data_list[0])
```

```python
def double_first_line(text):
    lines = text.split('\n')
    if len(lines) > 1:
        first_line = lines[0]
        new_text = first_line + '\n' + text
        return new_text
    else:
        return text
    
    
nv = double_first_line(data_list[0])
print(nv)

data_list_titredouble = []
for document in data_list:
    data_list_titredouble.append(double_first_line(document))
```

```python
data_tokens = []
for element in data_list_titredouble:
    tokens = word_tokenize(element)
    data_tokens.append(tokens)
    

print(data_tokens[0])
```

```python
data_stems = []
# Initialisation du lemmatiseur
lemmatizer = WordNetLemmatizer()
for element in data_tokens:
    data_stem = []
    for e in element:
        stemmed_word = lemmatizer.lemmatize(e)
        data_stem.append(stemmed_word)        
    data_stems.append(data_stem)

print(data_stems[0])
```

```python
filtered_stems = []
for element in data_stems:
    stop_words = set(stopwords.words('english'))
    filtered_stem = [word for word in element if not word.lower() in stop_words]
    filtered_stem = [s for s in filtered_stem if s not in string.punctuation]
    filtered_stem = [word.replace('"', '') for word in filtered_stem]
    filtered_stem = [word.replace("'", '') for word in filtered_stem]
    filtered_stem = [word.replace(".", '') for word in filtered_stem]
    filtered_stem = [word.replace("``", '') for word in filtered_stem]
    filtered_stems.append(filtered_stem)
print(filtered_stems[0])
```

#### a choice of indexing terms :  
using a combination of the most frequent and most discriminative terms can often provide the best results for information retrieval.

```python
# Flatten the list of lists into a single list
flat_stem_list = [stem for sublist in filtered_stems for stem in sublist]

# Count the occurrences of each stem
stem_counts = Counter(flat_stem_list)
del stem_counts['']

print(stem_counts)
```

Dans un premier temps je ne prends que les 10 premier termes

```python
top_10 = sorted(stem_counts.items(), key=lambda x: x[1], reverse=True)
top_10_words = [item[0] for item in top_10]

#TF PART

occurrences = []
# iterate over each document
for document in filtered_stems:
    # initialize a dictionary to store the counts of each word in the document
    word_counts = {word: 0 for word in top_10_words}
    
    # count the number of occurrences of each word in the document
    for word in document:
        if word in top_10_words:
            word_counts[word] += 1
    
    # append the counts to the list of occurrences
    occurrences.append(list(word_counts.values()))

# print the occurrences
# print(occurrences[0])
TF = []
i = 0
for element in occurrences:
    tfi = []
    for e in element:
        t = e/len(data_tokens[i])
        tfi.append(t)
    TF.append(tfi)
    i = i+1
#print(TF)


#IDF part 

# initialize a dictionary to store the counts of each word
word_counts = {word: 0 for word in top_10_words}

# iterate over each word
for word in top_10_words:
    # iterate over each document
    for doc in filtered_stems:
        # check if the word is in the document
        if word in doc:
            # increment the counter for that word
            word_counts[word] += 1

# print the number of documents containing each word
#for word, count in word_counts.items():
    #print(f'{word}: {count}')

words = [word for word, count in word_counts.items()]
counts = [count for word, count in word_counts.items()]

#print(words)
#print(counts)
i=0
IDF = []
for word in top_10_words:
    res = math.log(len(data_tokens) / counts[i])
    i = i+1
    IDF.append(res)

Weight = []
for element in TF:
    w = []
    for i, e in enumerate(element):
        res = pow(e,1) * pow(IDF[i],1.5)
        w.append(res)
    Weight.append(w)

#print(Weight)

for i, doc in enumerate(Weight):
    print(f"Document {i+1}:")
    for j, weight in enumerate(doc):
        if(weight != 0):
            print(f"{top_10_words[j]}: {weight:.3f}")
    print("\n")
```

```python
with open("DATA/EVALDS", "r") as f:
    queriess = f.read()
    print(queriess)
```

```python
queries_list = re.split(r'\n(?=.I \d+)', queriess)
queries_list = [re.sub(r'^\.I \d+\n', '', elem) for elem in queries_list]
print(queries_list[2])
```

```python
def process_document(document):
    #enlever le .T
    deb = document.split('.T')
    
    
    try : #le cas ou il y a tout les composant
        sd = deb[1].split('.A')
        titre = sd[0]

        suite = sd[1]

        th = suite.split('.W')

        auteur = th[0]


        suitee = th[1]

        futh = suitee.split('.B')

        contenu = futh[0]

        bibliographe = futh[1]
    except:
        try: #le cas ou il y a que le titre et le contenue
            th = deb[1].split('.W')
            titre = th[0]
            contenu = th[1]
        except: # le cas ou il y a que le contenue
            contenu = deb[0]
            titre = ""
    
    
    return titre + titre + contenu


print(process_document(queries_list[2]))
```

```python
queries = []
for document in queries_list:
    queries.append(process_document(document))
```

```python
query_tokens = []
for element in queries:
    tokens = word_tokenize(element)
    query_tokens.append(tokens)
    

print(query_tokens[2])
```

```python
queries_stems = []
for element in query_tokens:
    data_stem = []
    for e in element:
        stemmed_word = lemmatizer.lemmatize(e)
        data_stem.append(stemmed_word)        
    queries_stems.append(data_stem)

print(queries_stems[0])
```

```python
filtered_queries_stems = []
for element in queries_stems:
    stop_words = set(stopwords.words('english'))
    filtered_stem = [word for word in element if not word.lower() in stop_words]
    filtered_stem = [s for s in filtered_stem if s not in string.punctuation]
    filtered_stem = [word.replace('"', '') for word in filtered_stem]
    filtered_stem = [word.replace("'", '') for word in filtered_stem]
    filtered_stem = [word.replace(".", '') for word in filtered_stem]
    filtered_stem = [word.replace("``", '') for word in filtered_stem]
    filtered_queries_stems.append(filtered_stem)
print(filtered_queries_stems[0])
```

```python
TF_queries = [{word: count/len(word_list) for word, count in Counter(word_list).items()} for word_list in filtered_queries_stems]

print(TF_queries[7])
```

```python
tf_idf = []
for doc in TF_queries:
    tf_idf_doc = {}
    for i, word in enumerate(top_10_words):
        if word in doc:
            tf_idf_doc[word] = pow(doc[word],1) * pow(IDF[i],1.5)
        else:
            tf_idf_doc[word] = 0
    tf_idf.append(tf_idf_doc)

weights_queries = []
for e in tf_idf:
    weights_queries.append(list(e.values()))

print(weights_queries[7])
```

### Produit scalaire

```python
from tqdm import tqdm

ResultNormalisedQueries = []

for queryInteger in tqdm(range(len(weights_queries)), desc="Chargement"):
    result = [sum(a * b for a, b in zip(weights_queries[queryInteger], sublist)) for sublist in Weight]
    normeQuery = math.sqrt(sum([x**2 for x in weights_queries[queryInteger]]))
    normeDocumentList = []
    for document in Weight:
        normeDoc = math.sqrt(sum([x**2 for x in document]))
        normeDocumentList.append(normeDoc)

    resultNormalised = []
    i = 0
    for res in result:
        if normeQuery == 0:
            resultNormalised.append(0)
        else:
            resultNormalised.append(res / normeQuery * normeDocumentList[i])
        i = i + 1

    ResultNormalisedQueries.append(resultNormalised)
```

```python
for j in range(len(ResultNormalisedQueries)):
    top_10 = sorted(range(len(ResultNormalisedQueries[j])), key=lambda i: ResultNormalisedQueries[j][i])[:]
    for i in top_10:
        if ResultNormalisedQueries[j][i] > 0.075:
            print("Query ",j+1)
            print(f"Document: {i+1}, Valeur: {ResultNormalisedQueries[j][i]}")    
```

```python
with open("zakariyaa-elhaddioui_Hamza-Azeroual(run-name=[doubleTitre+IDF_a_plus_de_poids]).REL", "w") as fichier:
    for j in range(len(ResultNormalisedQueries)):
        top_10 = sorted(range(len(ResultNormalisedQueries[j])), key=lambda i: ResultNormalisedQueries[j][i])
        for i in top_10:
            if ResultNormalisedQueries[j][i] > 0.02:
                fichier.write(f"{j+1}\t{i+1}\t{ResultNormalisedQueries[j][i]}\t\n")
```
